```python
#importing the assets
import numpy as np
import cv2
```

```python
#reading the img
image = cv2.imread('/content/digits1.png')
```

```python
#converting the img to grayscale
gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
#dividing the img into 5000 small dimensions each of 20 x  20 pixels
divisioned_img = list(np.hsplit(i, 100)
for i in np.vsplit(gray_img, 50))
```

```python
from os import NGROUPS_MAX
#converting to numpy array
np_array= np.array(divisioned_img)
print(np_array)
```

```python
#prepping the training and testing data
train_data = np_array[:, :50].reshape(-1, 400).astype(np.float32)
testing_data = np_array[:,50:100].reshape(-1,400).astype(np.float32)
```

```python
#creating labels
k = np.arange(10)
train_labels = np.repeat(k, 250)[:, np.newaxis]
test_labels = np.repeat(k, 250)[:, np.newaxis]
```

```python
#initiate knn algo
knn = cv2.ml.KNearest_create()

#TRAINING
knn.train(train_data,
          cv2.ml.ROW_SAMPLE,
          train_labels
          )
```

```python
ret, output, neighbour, distance = knn.findNearest(testing_data, 3)
print(output)
print(neighbour)
print(distance)
```

```python
matched = output==test_labels
correct_op = np.count_nonzero(matched)
print(matched)
```

```python
#accuracy
acc = (correct_op * 100.0)/(output.size)
print(acc)
```
