```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import nltk
import io
import unicodedata
import re
import string
from numpy import linalg
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import wordnet
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
print("All libraries imported successfully.")
```

```python
with open('kindle.txt', encoding= 'ISO-8859-1') as f:
    text = f.read()

print("Text file read successfully.")
print(f"Length of text: {len(text)} characters.")
print(f"First 500 characters of text:\n{text[:500]}")
```

```python
#starting data preprocessing
#tokenization
sent_tokenizer = PunktSentenceTokenizer(text)
sentences_tokenised = sent_tokenizer.tokenize(text)
print('Sentences tokenised successfully.')
print(f"Number of sentences: {len(sentences_tokenised)}")
print(f'First 3 sentences:\n{sentences_tokenised[:3]}')
```

```python
# Ensure tokenizer resources are available for this environment
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

word_tokenized = word_tokenize(text)
print(word_tokenized[:10])
print(f"Total words tokenized: {len(word_tokenized)}")
```

```python
#porter stemmer
porter_stemmer = PorterStemmer()

nltk_tokens = nltk.word_tokenize(text)
print(nltk_tokens[:10])

#applying porter stemmer to the first 10 tokens
for word in nltk_tokens:
    stemmed_word = porter_stemmer.stem(word)
    print(f"Original word: {word} -> Stemmed word: {stemmed_word}")
```

```python
# lemmatization of words
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

lemmatizer = WordNetLemmatizer()
nltk_tokens = nltk.word_tokenize(text)

for word in nltk_tokens[:20]:
    lemmatized_word = lemmatizer.lemmatize(word)
    print(f"Original word: {word} --> Lemmatized word: {lemmatized_word}")
```

```python
# final part
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('universal_tagset', quiet=True)

with open('kindle.txt', encoding='ISO-8859-1') as f:
    text = f.read()

tokens = nltk.word_tokenize(text)
tagged_tokens = nltk.pos_tag(tokens, tagset='universal', lang='eng')
print(tagged_tokens[:30])
```

```python
# sentiment analysis
nltk.download('vader_lexicon', quiet=True)
nltk.download('punkt', quiet=True)

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
tokenizer = nltk.tokenize.load('tokenizers/punkt/english.pickle')
print("Sentiment analyzer initialized successfully.")

with open('kindle.txt', encoding='ISO-8859-1') as f:
    for text in f.read().split('\n'):
        print(text)
        scores = sid.polarity_scores(text)
        for key in sorted(scores):
            print(f"{key}: {scores[key]}")

    print()
```
