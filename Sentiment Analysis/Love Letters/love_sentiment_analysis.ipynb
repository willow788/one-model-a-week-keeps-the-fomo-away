```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import re
import string
import os
import io
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import wordnet
from nltk.corpus import webtext
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

print("All libraries imported successfully.")
```

```python
with open('love.txt', encoding='ISO-8859-1') as f:
    text = f.read()

print(text[:5000])
print('length of the entire document is: ', len(text))
```

```python
#starting data preprocessing
sentence_tokeniser = PunktSentenceTokenizer()
sentences = sentence_tokeniser.tokenize(text)
print(sentences[:5])
print('total length of the tokenized sentences is: ', len(sentences))
```

```python
#word tokenization
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

word_tokens = word_tokenize(text)
print(f'first 20 word tokens: {word_tokens[:20]}')
print('total no of word tokens: ', len(word_tokens))
```

```python
#stemming of words
porter_stemmer = PorterStemmer()
nltk_tokens = nltk.word_tokenize(text)

for word in nltk_tokens:
    stemmed_word = porter_stemmer.stem(word)
    print(f"Original word: {word}--> Stemmed word: {stemmed_word}")

    
```

```python
#snowball stemmer
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer('english')
nltk_tokens = nltk.word_tokenize(text)

for word in nltk_tokens:
    stemmed_by_olaf = snowball_stemmer.stem(word)
    print(f"Original word: {word}--> Stemmed word by snowball: {stemmed_by_olaf}")
    
```

```python
#lemmatization
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

lematizer = WordNetLemmatizer()
nltk_tokens = nltk.word_tokenize(text)

for word in nltk_tokens:
    lemmatized_word = lematizer.lemmatize(word)
    print(f"Original word: {word}--> Lemmatized word: {lemmatized_word}")
    
```

```python
#tagging of words
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('universal_tagset', quiet=True)

with open('love.txt', encoding='ISO-8859-1') as f:
    text = f.read()
tokens = nltk.word_tokenize(text)
tagged_tokens = nltk.pos_tag(tokens, tagset='universal', lang='eng')
length_of_tagged_tokens = len(tagged_tokens)
for i in range(length_of_tagged_tokens):
    print(tagged_tokens[i])
```

```python
#analysing someone else love letters, since i dont have any
nltk.download('punkt', quiet=True)
nltk.download('vader_lexicon', quiet=True)

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiment_analyzer = SentimentIntensityAnalyzer()
tokenizer = nltk.tokenize.load('tokenizers/punkt/english.pickle')
print("Sentiment analyzer initialized successfully.")

with open('love.txt', encoding= 'ISO-8859-1') as f:
    for text in f.read().split('\n'):
        print(text)
        scores = sentiment_analyzer.polarity_scores(text)
        for key in sorted(scores):
            print(f"{key}: {scores[key]}")

    print()
```
