```python
from sklearn import datasets
digits = datasets.load_digits()
dir(digits)
```

```python
#here the dir function was used to print the name of the features in the dataset
```

```python
print(digits.images[0])
```

```python
import matplotlib.pyplot as plt

def plot_a_dig(i):
  nplots = 16
  fig = plt.figure(figsize=(15, 15))
  for j in range(nplots):
    plt.subplot(4, 4, j+1)
    plt.imshow(digits.images[i+j], cmap= 'binary')
    plt.title(digits.target[i+j])
    plt.axis('off')

  plt.show()

plot_a_dig(0)
```

```python
x = digits.images
print(x)
```

```python
y = digits.target
x = digits.images.reshape((len(digits.images), -1))
#print(x)
x.shape
```

```python
#spliting the data into training and test sets
x_train = x[:1000]
y_train = y[:1000]
x_test = x[1000:]
y_test = y[1000:]
```

```python
#importing the multi layer perceptron
from sklearn.neural_network import MLPClassifier

#specifiying the details of the mlp classifiers
mlp_neuron  = MLPClassifier(hidden_layer_sizes= (15,),
                            activation = 'logistic',
                            alpha= 1e-4, solver= 'sgd',
                            tol= 1e-4, random_state = 1,
                            learning_rate= 'constant',
                            learning_rate_init= .1,
                            verbose= True)
```

```python
mlp_neuron.fit(x_train, y_train)
```

```python
#predicting on testing data
prediction = mlp_neuron.predict(x_test)
print(f'model prediction: {prediction[:50]}')

actual_labels = y_test[:50]
print(f'actual labels: {actual_labels}')

#printing the no of correct answers
```

```python
#importing accurracy score
from sklearn.metrics import accuracy_score
#importing confusion matrix
from sklearn.metrics import confusion_matrix

accuracy = accuracy_score(y_test, prediction)
print(f'model accuracy: {accuracy}')
print(f'model accuracy in percent: {accuracy * 100} %')
```

```python
#plotting tthe loss curve
fig, axes  = plt.subplots(1, 1)
axes.plot(mlp_neuron.loss_curve_, 'o-')
axes.set_xlabel('Epoch')
axes.set_ylabel('Loss')
axes.set_title('Loss curve')
plt.show()
```
