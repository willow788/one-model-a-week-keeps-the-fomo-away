```python
!pip install pytesseract
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt
print('all assets imported')
```

```python
img_path = cv2.imread('/content/vodka.png')
#converting img to rgb
img_rgb = cv2.cvtColor(img_path, cv2.COLOR_BGR2RGB)
#DISplaying the img once
plt.imshow(img_rgb)
plt.show()
```

```python
#converted to gray scale
img_grayed = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
print('Gray-Scaled Image')
plt.imshow(img_grayed, cmap='gray')
plt.show()
```

```python
#now we will extracttext from the image
text_extracted = pytesseract.image_to_string(img_rgb, lang='eng')
print(text_extracted)
```

```python
#drawing boxes around detected task
data_estracted = pytesseract.image_to_data(img_rgb, output_type=pytesseract.Output.DICT)

n_boxes = len(data_estracted['text'])
for box in range(n_boxes):
  (x, y, w, h) = (data_estracted['left'][box], data_estracted['top'][box], data_estracted['width'][box], data_estracted['height'][box])
  cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 0 , 0), 2)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.title('Detected Text are marked by rectangles')
plt.axis('off')
plt.show()
```
